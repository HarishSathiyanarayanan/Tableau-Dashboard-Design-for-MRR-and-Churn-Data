```python
#input the xls file
import pandas as pd
import numpy as np
```

```python
#read the xls file
df = pd.read_excel(r'C:\Users\haris\Desktop\MRR data.xlsx')
```

```python
#head of the data
df.head(20)
```

```python
#replace NaN with 'NA'
df = df.fillna('NA')
```

```python
#head of the data
df.head()
```

```python
# Parse the data to ensure the date column is in datetime format
df['Month and Year'] = pd.to_datetime(df['Month and Year'])

# Sort the data by Account ID and Month and Year
df = df.sort_values(by=['Account ID', 'Month and Year'])

# Get the starting ARR for each account
starting_arr = df.groupby('Account ID').first().reset_index()[['Account ID', 'ARR (in $)']]
starting_arr.columns = ['Account ID', 'Starting ARR']

# Get the ending ARR for each account
ending_arr = df.groupby('Account ID').last().reset_index()[['Account ID', 'ARR (in $)']]
ending_arr.columns = ['Account ID', 'Ending ARR']

# Merge starting and ending ARR with the original dataframe
df = df.merge(starting_arr, on='Account ID', how='left')
df = df.merge(ending_arr, on='Account ID', how='left')

# Calculate the GRR for each account
df['GRR'] = df.apply(lambda row: min(row['Ending ARR'], row['Starting ARR']) / row['Starting ARR'], axis=1)

df.head()
```

```python
df.head(60)
```

```python
# Calculate the NRR for each account
df['NRR'] = df.apply(lambda row: row['Ending ARR'] / row['Starting ARR'], axis=1)

df.head()
```
